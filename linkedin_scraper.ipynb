```python
import time 
import pandas as pd
import csv

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import NoSuchElementException
```

```python

# Find web elements/ input texts by XPath, CSS Selector, Id, ClassName, or TagName
def findByXPath(driver, XPATH):
    try:
        return WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, XPATH)))
    except Exception as e:
        print(e)
        #driver.quit()
        return None

def findBySelector(driver, selector):
    try:
        return WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, selector)))
    except Exception as e:
        #print(e)
        #driver.quit()
        return None
        
def findListBySelector(driver, selector):
    try:
        __ = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, selector)))
        items = driver.find_elements(By.CSS_SELECTOR, selector)
        list_items = []
        for x in items:
            item = x.text
            list_items.append(item) 
        return(list_items)
    except Exception as e:
        #print(e)
        #driver.quit()
        return None
   
        
def findById(driver, ID):
    try:
        return WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, ID)))
    except Exception as e:
        print(e)
        driver.quit()

def findByClassName(driver, class_name):
    try:
        return WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, class_name)))
    except Exception as e:
        print(e)
        #driver.quit()
        return None
        
        
def findByTagName(driver, tag_name):
    try:
        return WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, tag_name)))
    except Exception as e:
        print(e)
        #driver.quit()
        return None

def login(driver):
    EMAIL = 'nudata.lab@gmail.com'
    PASSWORD = "nudatalab2022"
    
    emailInput = findBySelector(driver,'#session_key')
    emailInput.send_keys(EMAIL)

    passInput = findBySelector(driver,'#session_password')
    passInput.send_keys(PASSWORD)
    
    submitButton = findByXPath(driver, '//*[@id="main-content"]/section[1]/div/div/form/button')
    submitButton.click()
    
def searchByName(driver,name):
    """
    Pass a name/ keyword to the search bar
    """
    searchInput = findByXPath(driver, '//*[@id="global-nav-typeahead"]/input')
    searchInput.click()
    searchInput.send_keys(name)
    searchInput.send_keys(Keys.ENTER)


# Set up chromedriver and log in
driver = webdriver.Chrome("C:/Users/nhatp/Desktop/Innsure/chromedriver.exe")
url = "https://www.linkedin.com/"
driver.get(url)
driver.maximize_window()
time.sleep(2)
login(driver)

# List of executives name
allProfiles = []
names = ["Michael Johnson Chief Information Security Meta", "LinkedIn CISO"]

for i in range(len(names)):
    searchByName(driver,names[i])
    keyword = (names[i])
    
    # click Peoples tab
    people = findBySelector(driver,"#search-reusables__filters-bar .search-reusables__filter-list li:nth-child(1)")
    people.click()

    #click the first profile
    getProfile = findBySelector(driver, '.entity-result__item:nth-child(1)')
    if not getProfile:
        continue
    getProfile.click()

    # get name
    getName = findByClassName(driver,"text-heading-xlarge")
    name = (getName.text)

    # get about
    getAbout = findBySelector(driver, '.pv-shared-text-with-see-more')
    about = getAbout.text
    time.sleep(2)


    # get number of connections
    getConnection = findByClassName(driver, f"pv-top-card--list-bullet")
    connection = getConnection.find_element_by_tag_name("li").text

    # get number of followers
    getFollower = findBySelector(driver,'#recent_activity ~.pvs-header__container .pvs-header__top-container--no-stack .pvs-header__left-container--stack .pvs-header__title-container p span:nth-child(1)')
    follower = getFollower.text

    # get location
    getLocation = findBySelector(driver, '.pb2 .text-body-small')
    location = getLocation.text

    # get education
    #print("Education:")
    education = findListBySelector(driver,"#education ~ .pvs-list__outer-container li span span:nth-child(1)")
    time.sleep(2)
        
    # get experience
    experience = findListBySelector(driver,"#experience ~ .pvs-list__outer-container li span span:nth-child(1)")

    # get licenses & certifications
    try:
        selector1 = "#licenses_and_certifications ~.pvs-list__outer-container .pvs-navigation__text"
        getCert_plus = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CSS_SELECTOR, selector1)))
    except:
        getCert_plus = None
        cert = findListBySelector(driver,"#licenses_and_certifications ~.pvs-list__outer-container ul li .display-flex.flex-column.full-width.align-self-center .display-flex.flex-row.justify-space-between .display-flex.flex-column.full-width span span:nth-child(1)")
        #print(certs)
    if getCert_plus is not None:
        getCert_plus.click()
        cert = findListBySelector(driver,"#main .display-flex.flex-column.full-width.align-self-center .display-flex.flex-row.justify-space-between .display-flex.flex-column.full-width span:nth-child(1) span:nth-child(1)")
        #print(certs)
        back1 = findBySelector(driver,"[aria-label='Back to the main profile page']")
        back1.click()
    time.sleep(2)
    
    # get all skills
        #check if there is a "Show more skills" button. If yes, click button and get all the skills.
        #if not, get all the skills available on profile page
    try:
        selector2 = "#skills ~.pvs-list__outer-container .pvs-navigation__text"
        getSk_plus = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CSS_SELECTOR, selector2)))
    except:
        getSk_plus = None
        skill = findListBySelector(driver,"#skills ~.pvs-list__outer-container ul li .display-flex.flex-column.full-width.align-self-center .display-flex.flex-row.justify-space-between .display-flex.flex-column.full-width span span:nth-child(1)")
        #print(skills)
    if getSk_plus is not None:
        getSk_plus.click()
        skill = findListBySelector(driver,"#main .display-flex.flex-column.full-width.align-self-center .display-flex.flex-row.justify-space-between .display-flex.flex-column.full-width span:nth-child(1) span:nth-child(1)")
        #print(skills)
        back2 = findBySelector(driver,"[aria-label='Back to the main profile page']")
        back2.click()
    profile = {'Name':name,'About':about,'Connection':connection,'Follower':follower,'Location':location,'Education':education,'Experience':experience,'License':cert,'Skills':skill}
    allProfiles.append(profile)

#driver.quit()

# get details for each profile    
df = pd.DataFrame(allProfiles)
df.to_csv('table.csv') 


driver.close()
```
